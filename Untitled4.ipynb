COVID-19 dataset (practice)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import covid-19 dataset
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid_df = pd.read_csv(url)

#1:exporole the dataset
print(covid_df.head())  #view first 5 rows
print(covid_df.info()) #info about dataset
print(covid_df.describe()) #statical summaray


#2:basic information
print(covid_df.shape)
print(covid_df['location'].unique)

#filtering data
#3:Find all data for a specific country
pakistan_data = covid_df[covid_df['location']=="Pakistan"]
print("\n PAKISTAN COVID-19 DATA",pakistan_data.head())
#Find days with more than 10,000 new cases globally
high_cases = covid_df[covid_df['new_cases'] > 10000]
print("\nHIGH CASES:",high_cases.head())

 #4:Aggregation and Grouping
#total number of cases for each country
total_deaths  = covid_df.groupby('location')['total_deaths'].max()
print("\n TOTAL DEATHS",total_deaths)
#average daily vaccination rate in the top 5 countries with the most vaccinations
top_vaccinations = covid_df.groupby('location')['total_vaccinations'].max().sort_values(ascending = False).head(5)
print("\nTOP VACCINATIONS:",top_vaccinations)
top_vaccinated_countries = covid_df[covid_df['location'].isin(top_vaccinations.index)]
print("Top vaccinated countries:",top_vaccinated_countries)
avg_daily_vaccinations = top_vaccinated_countries.groupby('location')['new_vaccinations'].mean()
print("\nAverage daily vaccinations:",avg_daily_vaccinations)


#5: visualization of covid-19 dataset
#Plot the trend of new cases in a specific country
pakistan_data.plot(x='date', y='new_cases', title='New Cases In Pakistan', kind='line')
plt.show()
#Compare total deaths in the top 5 countries with the most cases:
top_cases = covid_df.groupby('location')['total_cases'].max().sort_values(ascending=False).head(5)
top_deaths = covid_df[covid_df['location'].isin(top_cases.index)].groupby('location')['total_deaths'].max()
top_deaths.plot(kind='bar', title="Totel Deaths In Top 5 Countries")
plt.show()


```
